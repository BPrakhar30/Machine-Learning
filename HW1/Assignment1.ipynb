```python
def includes(item, val, start_ind=None): 

   if start_ind == None:               
         # initialising start index = 0 if the value is not specified 
    start_ind = 0

   k=0                                                  
      # When a list is entered
   if(isinstance(item, list)):
    for i in range(start_ind, len(item)):
      if(item[i]==val): 
        k=1
    if(k==1):
      print(True) 
    else: 
      print(False) 

   elif(isinstance(item, str)):                           
       # When a string is entered
    for i in range(start_ind, len(item)): 
      if(item[i]==val): 
        k=1 
    if (k==1): 
        print(True)  
    else: 
        print(False) 
   else:
     if val in item.values():
       print(True) 
     else: 
       print(False) 
```

```python
includes([2, 3, 4], 2, 0) # True 
includes([2, 3, 4], 2, 1) # False 
includes([2, 3, 4], 4, 1) # True 
includes({'a':1,'b':2}, 1) # True 
includes({'a':1,'b':2}, 'a') # False
includes('abcd', 'b') # True 
includes([2.4, 7.5, -9.6], 2.4, 1) #False 
```

```python
import numpy as np
import matplotlib.pyplot as mplt

def predict(x, B_0, B_1):
    return B_0 + np.multiply(B_1, x)

def MSE(y , y_pred):
    return (np.mean(np.power(y- y_pred, 2)))

def compute_loss_gradient(x, y, y_pred):
    D_B_0 = (-2) * np.mean(x.dot(y - y_pred))
    D_B_1 = (-2) * np.mean(y - y_pred)
    return D_B_0, D_B_1
```

```python
def gradient_descent(x, y, B_0, B_1, num_epochs, lr):
    B_0_list = []
    B_1_list = []
    loss_list = []
    for i in range(num_epochs):
      B_0_list.append(B_0)
      B_1_list.append(B_1)

      y_pred = predict(x, B_0, B_1)
      loss = MSE(y, y_pred)
      loss_list.append(loss)
      DB_0, DB_1 = compute_loss_gradient(x, y, y_pred)
      B_1 = B_1 - lr * DB_1
      B_0 = B_0 - lr * DB_0
    
    return B_0_list, B_1_list, loss_list

data = np.load('p3_data.npy')
x=data[:,0]
y=data[:,1]

plt.figure(figsize=(6,4))
plt.scatter(x, y, s=25, c='black')
plt.show()

B_0_list, B_1_list, loss_list = gradient_descent(x, y, 0.2, 1.4, 100, 0.0001)
y_pred = predict(x, B_0_list[-1], B_1_list[-1])

plt.figure(figsize=(6,4))
plt.scatter(x, y, s=25, c='black')
plt.plot([min(x), max(x)], [min(y_pred), max(y_pred)], color='red')
plt.show()
print("The value of B_0 and B_1: ",B_0_list[-1],B_1_list[-1])
print("Loss Obtained: ",loss_list[-1])

print("Final Loss is: ", loss_list[-1])
plt.figure(figsize=(6, 4))
plt.plot(loss_list)
plt.xlabel("Epoch = Iteration")
plt.ylabel("Loss")
plt.title("Loss vs. Iteration")
plt.show()
```

```python
for lr in [0.0002, 0.0006, 0.01, 10]:
    B_0_list, B_1_list, loss_list = gradient_descent(x, y, 0.2, 1.4, 50, lr)
    print(f"final B_0, B_1 for learning rate {lr} is: ", B_0_list[-1], " ", B_1_list[-1])
    plt.figure(figsize=(15,4))
    ax1 = plt.subplot(1,2,1)
    ax1.plot(loss_list)
    ax1.set_xlabel("epoch")
    ax1.set_ylabel("loss")
    plt.title(f"LR is {lr}")
    plt.show()
```

When the learing rate(alpha) increases the funtion tends to diverge instead of getting converged. Therefore, the best learning rate would be 0.0002 as the loss function converges.

```python
import numpy as np 
import matplotlib.pyplot as plt 

def predict(x, B_0, B_1): 
    return B_0 + np.multiply(B_1, x)

def MSE(y, y_pred):
    return (np.mean(np.power(y - y_pred, 2)))

def compute_loss_gradient(x, y, y_pred):
   D_B_0 = (-2) * np.mean(x.dot(y - y_pred)) 
   D_B_1 = (-2) * np.mean(y - y_pred) 
   return D_B_0, D_B_1
```

```python
def Mini_GD(x, y, B_0, B_1, num_epochs, lr, batch_size):
  B_0_list = []
  B_0_list = []
  loss_list = []
  num_batches = len(x) // batch_size
  num_epochs = num_epochs // num_batches
  for i in range(num_epochs):
    for b in range(num_batches):
      B_0_list.append(B_0)
      B_1_list.append(B_1)
      y_pred = predict(x, B_0, B_1)
      loss = MSE(y, y_pred)
      loss_list.append(loss)
      D_B_0, D_B_1 = compute_loss_gradient(x, y, y_pred)
    B_1 = B_1 - lr * D_B_1
    B_0 = B_0 - lr * D_B_0
  return B_0_list, B_1_list, loss_list

data = np.load('p3_data.npy')
x= data[:,0]
y= data[:,1]

plt.figure(figsize=(6,4))
plt.scatter(x, y, s=25, c='black')
plt.show()

B_0_list, B_1_list, loss_list = Mini_GD(x, y, 0.2, 1.4, 100, 0.0001, 20)
y_pred = predict(x, B_0_list[-1], B_1_list[-1])
plt.figure(figsize=(6,4))
plt.scatter(x,y, s=25, c='black')
plt.plot([min(x), max(x)], [min(y_pred), max(y_pred)], color='red')
plt.show()
print("The Value of B_0 and B_1: ", B_0_list[-1], B_1_list[-1])
print("Loss Obtained: ", loss_list[-1])
```

```python
print("Final Loss is equal to", loss_list[-1])
plt.figure(figsize=(6,4))
plt.plot(loss_list)
plt.xlabel("Epoch = Iteration")
plt.ylabel("Loss")
plt.ylabel("Loss vs. Iteration")
plt.show()
```

As we have used the mini batch method here with the batch size of 20 and the number of iterations as 100 the loss is changed every 5(100%20) iterations.
